# encoding categorical data to numerical data
* label encoding is used as we have ordial data in salary.

```python
import category_encoders as ce
import pandas as pd
```

```python
# create object of Ordinalencoding
encoder= ce.OrdinalEncoder(cols=['salary'],return_df=True,
                           mapping=[{'col':'salary',
'mapping':{'low':0,'medium':1,'high':2}}])
```

```python
encoder
```

```python
y_encoded = encoder.fit_transform(y)
```

```python
y_encoded.nunique()
```

```python
y_encoded
```

```python
encoder=ce.OneHotEncoder(cols='Department',handle_unknown='return_nan',return_df=True,use_cat_names=True)
```

```python
X_encoded = encoder.fit_transform(X)
```

```python
X_encoded
```

# now doing train_test_split 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_encoded,y_encoded,test_size = 0.2 ,random_state=42)
```

```python
y_train.nunique()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
model.fit(X_train, y_train)
model.score(X_test, y_test) # accuracy
```
